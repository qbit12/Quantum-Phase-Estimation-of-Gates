```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from numpy import pi

from qiskit import QuantumCircuit,ClassicalRegister,QuantumRegister,Aer,execute
from qiskit.visualization import plot_histogram
```

```python
qpe = QuantumCircuit(5,4)
qpe.x(4)
for i in range(4):
    qpe.h(i)
```

```python
repetitions = 1
for counting_qubit in range(4):
    for i in range(repetitions):
        qpe.cu1(np.pi/4, counting_qubit, 4);
    repetitions *= 2                              #check qiskit textbook for more details
qpe.draw(output='mpl')
```

```python
def qft_dagger(circ, n):
    for qubit in range(n//2):
        circ.swap(qubit, n-qubit-1)
    for j in range(n):
        for m in range(j):
            circ.cu1(-np.pi/float(2**(j-m)), m, j)
        circ.h(j)
```

```python
qpe.barrier()
qft_dagger(qpe, 4)
qpe.barrier()
for n in range(4):
    qpe.measure(n,n)
```
